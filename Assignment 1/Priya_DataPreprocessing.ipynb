```python
import pandas as pd
```

```python
# reads csv files in data frames
DataSet1=pd.read_csv("FirstSetQueryResults.csv")
DataSet2=pd.read_csv("SecondSetQueryResults.csv")
DataSet3=pd.read_csv("ThirdSetQueryResults.csv")
DataSet4=pd.read_csv("FourthSetQueryResults.csv")
```

```python
from bs4 import BeautifulSoup
```

```python
# function to remove html tags
def removetags(raw_data):
    cleandata=BeautifulSoup(raw_data, "lxml").text
    return cleandata
```

```python
# data frame before removing html tags
DataSet1.head()
```

```python
# removing html tags from body column
DataSet1["Body"]=DataSet1["Body"].apply(removetags)
DataSet1.head()
```

```python
DataSet2["Body"]=DataSet2["Body"].apply(removetags)
DataSet2.head()
```

```python
DataSet3["Body"]=DataSet3["Body"].apply(removetags)
DataSet3.head()
```

```python
DataSet4["Body"]=DataSet4["Body"].apply(removetags)
DataSet4.head()
```

```python
import re
import string
```

```python
# function to remove punctuation
def removepuctuations(data):
    pattern = r'[' + string.punctuation + ']'
    return re.sub(pattern, '', data)
```

```python
# removing punctuations from body column
DataSet1["Body"]=DataSet1["Body"].apply(removepuctuations)
DataSet2["Body"]=DataSet2["Body"].apply(removepuctuations)
DataSet3["Body"]=DataSet3["Body"].apply(removepuctuations)
DataSet4["Body"]=DataSet4["Body"].apply(removepuctuations)
```

```python
# removing punctuations from title column
DataSet1["Title"]=DataSet1["Title"].apply(removepuctuations)
DataSet2["Title"]=DataSet2["Title"].apply(removepuctuations)
DataSet3["Title"]=DataSet3["Title"].apply(removepuctuations)
DataSet4["Title"]=DataSet4["Title"].apply(removepuctuations)
```

```python
DataSet1 = DataSet1.replace(r'\n',' ', regex=True) 
DataSet1 = DataSet1.replace(r'\t',' ', regex=True)
DataSet1 = DataSet1.replace(r'\r',' ', regex=True)
DataSet1 = DataSet1.replace(r'\b',' ', regex=True)
DataSet1 = DataSet1.replace(r'\f',' ', regex=True)

DataSet2 = DataSet2.replace(r'\n',' ', regex=True) 
DataSet2 = DataSet2.replace(r'\t',' ', regex=True)
DataSet2 = DataSet2.replace(r'\r',' ', regex=True)
DataSet2 = DataSet2.replace(r'\b',' ', regex=True)
DataSet2 = DataSet2.replace(r'\f',' ', regex=True)

DataSet3 = DataSet3.replace(r'\n',' ', regex=True) 
DataSet3 = DataSet3.replace(r'\t',' ', regex=True)
DataSet3 = DataSet3.replace(r'\r',' ', regex=True)
DataSet3 = DataSet3.replace(r'\b',' ', regex=True)
DataSet3 = DataSet3.replace(r'\f',' ', regex=True)

DataSet4 = DataSet4.replace(r'\n',' ', regex=True) 
DataSet4 = DataSet4.replace(r'\t',' ', regex=True)
DataSet4 = DataSet4.replace(r'\r',' ', regex=True)
DataSet4 = DataSet4.replace(r'\b',' ', regex=True)
DataSet4 = DataSet4.replace(r'\f',' ', regex=True)
```

```python
# merged all 4 files using concat
MergedData = pd.concat([DataSet1,DataSet2,DataSet3,DataSet4])
```
